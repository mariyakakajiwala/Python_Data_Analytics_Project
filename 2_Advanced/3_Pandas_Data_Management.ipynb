```python
print(id(df_original))
id(df_altered)

# since they both have the same ID, they're both basically the same df with 2 names/variables. So changes made to 1 will automatically affect the other.
# Both variables are referencing the same dataframe. However, we don't want that to happen cause that's not good actually.
# Therefore it's better to use the copy method.
```

# Using Copy() Method

```python
dataset = load_dataset('lukebarousse/data_jobs')
df_original = dataset['train'].to_pandas()
```

```python
df_altered = df_original.copy()
```

```python
median_salary = df_altered['salary_year_avg'].median()

df_altered['salary_year_avg'] = df_altered['salary_year_avg'].fillna(median_salary)

df_altered.loc[:5, 'salary_year_avg']
```

```python
df_original.loc[:5, 'salary_year_avg']
```
