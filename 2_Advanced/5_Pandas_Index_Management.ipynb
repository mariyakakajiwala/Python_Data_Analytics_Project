```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.sample(5)
```

```python
df.index
# This accesses the index attribute
```

```python
df.index.name
# This is asking for the name of the index column. If it has no name, u can assign it one.
```

```python
df.index.name = 'job_index'

df.head(5)
```

```python
df.index.dtype
```

# 3 Methods related to the Index
- df.reset_index()
- df.set_index()
- df.sort_index()

```python
df_usa = df[df['job_country'] == 'United States']
```

```python
df_usa.head(5)
```

```python
df_usa.reset_index(inplace=True)

df_usa
```

```python
df_usa.set_index('job_index', inplace=True)

# The set_index method is used to set the index for your dataframe according to a specified column
# It may also drop data at times (just as it dropped the other index column which we had previously made using reset_index)
```

```python
df_usa
```

```python
median_pivot = df_usa.pivot_table(values='salary_year_avg', index='job_title_short', aggfunc=['median', 'min', 'max'])
```

```python
median_pivot
```

```python
# We can use the median and the salary_year_avg columns to sort the values using sort_values

median_pivot.sort_values(by=('median', 'salary_year_avg'), ascending=False)
```

```python
# Using sort_index to sort the values
median_pivot.sort_index(inplace=True)
# This sorts the first column, that is of job_title_short (which is the index in this case) in ascending/alphabetical order
```

```python
median_pivot
```
